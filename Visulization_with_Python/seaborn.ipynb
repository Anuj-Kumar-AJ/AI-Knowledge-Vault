### Different types of distribution plots

1. Distribution Plot

This plot is for one variable or column ( this is an histogram plot, with kernal density estimator on top of histogram) 
kde is bassically probability density estimator

```python
# plot with kernal density estimator
sns.distplot(crash_df['not_distracted'])

# plot without kernal density estimator
sns.displot(crash_df['not_distracted'], kde=False)

# to plot data with different interval (here called bin)

sns.displot(crash_df['not_distracted'], kde=False, bins= 5)
```

2. Joint Plot
   
It is used to plot 2 distribution on X and Y axis
it used scatter plot

Here kind argument is used to plot differenty types od curves between both variable, like regression curve (`regression`), kde(`kde`) curve, hexagonal curve (`hex`)

it needed 4 argument
x, y, dataframe, and types of curve 

```python
sns.jointplot(x='speeding', y ='alcohol',data=crash_df, kind='kde', )
```

3. KDE Plot

kde -> kernal density estiator , here kernal is basically gaussion kernal in seaborn

```python
sns.kdeplot(crash_df['alcohol'])
```

4. Pair Plots

It plots relation between entire dataframe numerical values.
It gives corelation between all the columns .

To check for categorical dataset. use `hue`

```python
sns.pairplot(crash_df)

tips_df = sns.load_dataset('tips')

print(tips_df)

# hue is use to load categorical data 
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

5. Rug Plot

It creates sticks like lines , and just like in histogram , more density meaning higher histogram bar. here in rug plot more number meaning, denser color , and less number meaning ligher color

```python
sns.rugplot(tips_df['tip'])
```

6. Styling

```python
# sets the background style of plot
sns.set_style('dark')

# to change the size of plot
plt.figure(figsize=(8,4))

# to change the labels and texts size

sns.set_context('notebook', font_scale=1.4)

# to remove axis visibility in the plot 
sns.despine(left=False)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

### Categorical Plots

we check the distribution of categorical data in refrence to oneof the numeric data

1. Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.mean)

# here estimator is how we aggragate our data, meaning how its based on mean or std or median .. Default is mean
```

2. Count Plots

similar to barplot , but the estimator will count the number of occurances .

```python
sns.countplot(x='sex', data=tips_df)
```

3. Box Plots

compare the value of different variable, it shows us coartile of the data

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

# to change the legend position

plt.legend(loc=0)
```

4. Violin Plot

Its combination of box plot and kde plot


```python
sns.violinplot(x='day', y='total_bill', data=tips_df ,hue='sex')
```

5. Strip Plot

draw all the datapoint, where one variable is categorical

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)

# jitter -> make data spread out a bit so it wont stack top of one another
# hue -> spaerate sex with different color
#dodge -> make saperate colums for differnet sex
```

6. Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data = tips_df)
```

7. Palettes

To make plots looks nicer by using different colors using color maps. using palettes


```python
# to change figure size
plt.figure(figsize=(8,5))

# to chaneg looks of the plot
sns.set_style('dark')

# to change the size of text written on the plot
sns.set_context('talk')

sns.stripplot(x='day', y='total_bill', data=tips_df, hue= 'sex', palette='magma')
```

### Matrix Plot

1. Heatmaps, 

IT give info about coorelation between differnt variable or features, using different colors

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

# how crash df looks like
##print(crash_df.head())

# creating coorelation between differnet columns and storing it in crash matrix variable
crash_matrix = crash_df.corr()
print(crash_matrix.head())

# creating heatmap

sns.heatmap(crash_matrix,annot=True, cmap='Blues')

#annot -> add annotation( values), cmap is color map
```

creating matrix using pivit table


```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month', columns='year', values='passengers')
flights

# heatmap creation
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidths=1)
```

2. Cluster map

it creates cluster of nearest points

```python
iris = sns.load_dataset('iris')
print(iris)

species = iris.pop('species')
sns.clustermap(iris)
```

```python
sns.clustermap(flights, cmap='Blues', standard_scale=1)
```

3. Pair grid

It gives more control of perticular plot, which comes out of `pairplot`

```python
iris = sns.load_dataset('iris')
iris_q = sns.PairGrid(iris, hue='species')

iris_q.map(plt.scatter)

#iris_q.map_diag(plt.hist) # add histogram in diagonal

# upper part use scatter plot, and lower part use kde plot
iris_q.map_upper(plt.scatter)
iris_q.map_lower(sns.kdeplot)
```

4. Facet Grid

can plot modular plot in grid, we can draw custom plot inside it

```python
# tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
# tips_fg.map(plt.hist, 'total_bill')


# 2nd plot

tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=1.4, col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```
